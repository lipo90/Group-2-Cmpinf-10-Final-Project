A section about saftey is how many guns are being seized. Less guns being seized would indicate that there is less illegal activity involving guns.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
firearm = pd.read_csv("firearmsSeizures.csv", sep=",")
firearm
```

```python
byHood = firearm.groupby(['neighborhood'],as_index=False)['total_count'].sum()
byHood
```

```python
byHood.nsmallest(10, 'total_count').plot.bar(x = 'neighborhood', y = 'total_count', rot =80)
plt.show()
```

```python
byHood.nsmallest(10, 'total_count')
```

```python
byHoodandYear = firearm.groupby(['neighborhood', 'year'],as_index=False)['total_count'].sum()
byHoodandYear
```

```python
byHoodin22 = firearm[firearm['year'] ==2022].groupby(['neighborhood', 'year'],as_index=False)['total_count'].sum()
byHoodin22
```

```python
byHood.nsmallest(15, 'total_count')
```

```python
byHood.nlargest(10, 'total_count')
```

```python
byHoodin22.nsmallest(15, 'total_count')
```
