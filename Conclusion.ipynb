# The Best Neighborhood in Pittsburgh

## Team Procrastination

### Introduction

Using the datasets from the WPRDC, we sought the best neighborhood by comparing:
1. The fire incidents throughout the neighborhoods
2. The amount of crimes committed throughout the neighborhoods
3. The amount of firearm seizures throughout the neighborhoods

### The Metric

Our metric used for determining the "best" neighborhood was safety. A higher number of fire incidents indicates substantially less safety, as fire is harmful. The number of crimes was examined because More Crime = Less Safety, and a higher amount of firearm seizures throughout a neighborhood indicates people are misusing firearms, which is not safe.

### Fire Incidents

For Fire Incidents, we examined which neighborhood had the least amount of fires throughout. Simply put, having less fires make a neighborhood safer, as fire is destructive and deadly.

Winner: Mount Oliver Borough

Loser: Carrick

A close second to last place goes to South Side Flats, which is important for later

### Crimes Committed

Another factor considered was how many crimes were commited. More crimes make a neighborhood generally less safe. Some crimes are worse, but they show an overall trend towards unsafety the more there are.

Winner: Saint Clair

Loser: Cental Business District

A close second to last place, again, goes to South Side Flats. I'm sure you can see the theme here.

### Firearm Seizures

Firearm Seizures was the last factor we considered. In vein with the amount of crimes committed, the higher the amount of firearm seizures, the less safe a neighborhood is. It shows a general lack of care employed when handling firearms, which is dangerous for the average civilian.

Winner: Summer Hill/Swisshelm Park

Loser: Homewood South

You will not believe which neighborhood is in second to last place. It is, in fact, South Side Flats.

### Conclusion

Overall, the the numbers show a general trend of good neighborhoods. Ultimately, we decided upon Saint Clair as being the best neighborhood, as it has the least amount of crime, a very low amount of fire incidents, and a fairly low amount of firearm seizures. As for a loser, it has to be South Side Flats, as they have high crimes rates, fires, and firearm seizure rates, landing in second to last place in every factor considered.
